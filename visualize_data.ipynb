```python
df_v1cross_1_300_300 = \
pd.read_csv("books_results/DCN_V1/curve_plot_1crossLayers__[300, 300]_deepLayers[]concatLayers_20epochs1e-05LR/results.csv")


df_v2cross_1_300_300 = \
pd.read_csv("books_results/DCN_V2/curve_plot_1crossLayers__[300, 300]_deepLayers[]concatLayers_20epochs1e-05LR/results.csv")

```

```python
df_loss = pd.DataFrame({'Epochs': df_cross_0_300_500_300['EPOCHS'], 
                   'Standard DeepNet, 0 Cross Layers, [300, 500, 300]': df_cross_0_300_500_300['Validation Loss'], 
                   'DCN V1, 1 Cross Layers [300, 300]': df_v1cross_1_300_300['Validation Loss'],
                 'DCN V2, 1 Cross Layers [300, 300]': df_v2cross_1_300_300['Validation Loss']})

df_accuracy = pd.DataFrame({'Epochs': df_cross_0_300_500_300['EPOCHS'], 
                   'Standard DeepNet, 0 Cross Layers, [300, 500, 300]': df_cross_0_300_500_300['Validation Accuracy'], 
                   'DCN V1, 1 Cross Layers [300, 300]': df_v1cross_1_300_300['Validation Accuracy'],
                 'DCN V2, 1 Cross Layers [300, 300]': df_v2cross_1_300_300['Validation Accuracy']})
```

```python
df_melted_loss = df_loss.melt(id_vars=['Epochs'], var_name='Architecture', value_name='Loss')
df_melted_accuracy = df_accuracy.melt(id_vars=['Epochs'], var_name='Architecture', value_name='Accuracy')

```

```python
#sns.set_style("whitegrid")
sns.set_theme(style='dark')
```

```python
pd.option_context('mode.use_inf_as_na', True)
```

```python
with pd.option_context('mode.use_inf_as_na', True):
    # Create a line plot using Seaborn
    sns.lineplot(data=df_melted_loss, x='Epochs', y='Loss', hue='Architecture')
    # Show the plot
    plt.title('Loss vs Epochs.')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.show()
```

```python
with pd.option_context('mode.use_inf_as_na', True):
    # Create a line plot using Seaborn
    sns.lineplot(data=df_melted_accuracy, x='Epochs', y='Accuracy', hue='Architecture')
    # Show the plot
    plt.title('Loss vs Epochs.')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.show()
```
